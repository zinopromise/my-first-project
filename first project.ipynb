```python
reviews_max = {}
for app in andriod:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
print(len(reviews_max))
```

```python
andriod_clean = []
already_added = []
for app in andriod:
    name = app[0]
    n_reviews = float(app[3])
    if n_reviews == reviews_max[name] and name not in already_added:
        andriod_clean.append(app)
        already_added.append(name)
```

We are to develop an app that would be succcessful on both the App store and the Google Playstore. To do this, we need to find app profiles that are successful on both markets. Any app profile that works on both markets is more likely to attract both the Playstore and App store users, thereby making the app mpre successful. Our validation strategy is:
1. Build a minimal Andriod version of the app and add it to play store. This is because we have more app data from the Playstore than the App store.
2. If the app has a good response from users, develop it further. This is needed to keep existing users and also attract more new users.
3. If the app is profitable after 6 months, we build an IOS version of the app and add it to App store. This 6 months can be referred to as the probabtion period.

To generate a frequency table for the Google Play Store apps, we would need: App, Category, Genres.
For the ios Appstore, we would need: Track_name, prime_genre.

```python
def freq_table(dataset, index):
    freq = {}
    total = 0
    for app in dataset:
        total += 1
        value = app[index]
        if value in freq:
            freq[value] += 1
        else:
            freq[value] = 1
    percent = {}
    for key in freq:
        percentage = (freq[key]/ total) * 100
        percent[key]= percentage
    return percent

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
        
print(display_table(andriod_free, 1))
print('\n')
print(display_table(andriod_free, 9))
print('\n')
print(display_table(ios_free,11))
```

```python
for genre in freq_table(ios_free, -5):
    total = 0
    len_genre = 0
    for app in ios_free:
        genre_app = app[-5]
        if genre_app == genre:
            numb = float(app[5])
            total += numb
            len_genre += 1
    avg_user_rat = total / len_genre
    print(genre, ':', avg_user_rat)
```

In th code above, we calculated the average number of installs for each genre of the ios app data. We take the total number of user ratings as a proxy which can be find in the rating_count_tot column.

From the above findings, my app genre recommendation for the ios App store is Navigation, Reference and Business.
